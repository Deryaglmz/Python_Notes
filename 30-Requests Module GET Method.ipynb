```python
from collections import Counter

class CrimeReport():
    def __init__(self, regions, history, crime_type='all-crime'):
        self.regions = regions
        self.history = history
        self.crime_type = crime_type
        self.crimes = self.find_the_crime()

        
    def find_the_crime(self):
        crime_URL = 'https://data.police.uk/api/crimes-no-location'
        
        payload = {
        'category' : self.crime_type,
        'force' : self.regions,
        'date' :  self.history}
        response = requests.get(crime_URL, params = payload )
        
        if response.status_code == 200:
            return response.json()
        else :
            return None
        
    def report_offenders(self):
        list_of_criminals = []
        
        if self.crimes is not None :
            for crime in self.crimes:
                list_of_criminals.append(crime['category'])
                
            return Counter(list_of_criminals)    
        
```

```python
sr = CrimeReport('city-of-london', '2023-01' )
```

```python
sr.report_offenders()
```

```python
graphic_URL = 'https://image-charts.com/chart'
# https://image-charts.com/chart?chan&chd=t%3A60%2C40&chf=ps0-0%2Clg%2C45%2Cffeb3b%2C0.2%2Cf44336%2C1%7Cps0-1%2Clg%2C45%2C8bc34a%2C0.2%2C009688%2C1&chl=Hello%7CWorld&chs=700x190&cht=p3
payload = { 
    'chs' : '700x190',
    'chd' : 't:60,40',
    'cht' : 'p3',
    'chl' : 'Hello|World',
    'chan' : None ,
    'chf' : 'ps0-0,lg,45,ffeb3b,0.2,f44336,1|ps0-1,lg,45,8bc34a,0.2,009688,1'
}

response = requests.post(graphic_URL, data=payload)
```

# Radar Chart

```python
graphic_URL = 'https://image-charts.com/chart'

payload = { 
    'chco' : '3092de',
    'chd' : 't:81,65,50,67,59,81',
    'chl' : 'speed|smash|rust|ball_drift|defender|physique',
    'chdl' : 'Falcao',
    'chdlp' : 'b',
    'chs' : '480x480',
    'cht' : 'r',
    'chtt' : 'football player traits'
}

response = requests.post(graphic_URL, data=payload)

image = Image.open(BytesIO(response.content))
image.show()
```

```python
class footballer():
    def __init__(self, name, speed, smash, rust, ball_drift, defender, physique ):
        self.name = name
        self.speed= speed
        self.smash= smash
        self.rust = rust
        self.ball_drift = ball_drift
        self.defender= defender
        self.physique = physique
        
        
    def prepare_talent(self):
        return ','.join([
            str(self.speed),
            str(self.smash),
            str(self.rust),
            str(self.ball_drift),
            str(self.defender),
            str(self.physique),
            str(self.speed)
       ]) 
        
    def visualize_talent(self):
        
        graphic_URL = 'https://image-charts.com/chart'
        
        payload ={
            
            'chco' : '3092de',
            'chd' : 't' + self.prepare_talent(),
            'chdl' : self.name,
            'chdlp' : 'b',
            'chs' : '480x480',
            'chdlp' : 'b',
            'cht' : 'r',
            'chtt' : 'football player traits',
            'chl' : 'speed|smash|rust|ball_drift|defender|physique',
            'chxl': '0: 0|20|40|60|80|100',
            'chxt': 'x',
            'chxr': '0,0.0,100.0',
            'chm': 'B,AAAAAABB,0,0,0',
        }
        response = requests.post(graphic_URL, data=payload)

        image = Image.open(BytesIO(response.content))
        image.show()
        
        
    def show_talent_comparison(self, target_football_player):
        
        graphic_URL = 'https://image-charts.com/chart'
            
        payload ={
              
             
                'chco' : '3092de, 027182',
                'chd' : 't' + self.prepare_talent() + '|' +  target_football_player.prepare_talent(),
                'chdl' : self.name  + '|' + target_football_player.name ,
                'chdlp' : 'b',
                'chs' : '480x480',
                'chdlp' : 'b',
                'cht' : 'r',
                'chtt' : 'football player traits',
                'chl' : 'speed|smash|rust|ball_drift|defender|physique',
                'chxl': '0: 0|20|40|60|80|100',
                'chxt': 'x',
                'chxr': '0,0.0,100.0',
                'chm': 'B,AAAAAABB,0,0,0|B,0073CFBB,1,0,0'
          }
         
        response = requests.post(graphic_URL, data=payload)

        image = Image.open(BytesIO(response.content))
        image.show()
        
        
```

```python
messi = footballer('Messi ',85 ,92 ,91 ,95 ,38 ,65)
ronaldo= footballer('Ronaldo ',89 ,91 ,83 ,95 ,35 ,77)
```

```python
messi.visualize_talent()
```

```python
messi.show_talent_comparison(ronaldo)
```
